```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv('netflix_data.csv')
netflix_df.head()

movies_df = netflix_df[netflix_df['type'] == 'Movie']


movies_1990s = movies_df[
    (movies_df['release_year'] >= 1990) & (movies_df['release_year'] < 2000)
]
movies_1990s.head()

plt.hist(movies_1990s['duration'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()


# Some datasets use 'listed_in' instead of 'genre'
if 'genre' in movies_1990s.columns:
    genre_col = 'genre'
else:
    genre_col = 'listed_in'

action_movies_1990s = movies_1990s[movies_1990s[genre_col].str.contains('Action', na=False)]
action_movies_1990s.head()

short_action_movies = action_movies_1990s[action_movies_1990s['duration'] < 90]
short_movie_count = short_action_movies.shape[0]

print(f"Number of Action movies under 90 minutes in the 1990s: {short_movie_count}")


```
